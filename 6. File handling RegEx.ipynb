## Regular Expression - REgex

* match
* search
* findall
* finditer

```python
text = '''bcdefghijklmnopqurtuvwxyz
ABCDEFGHIJKLMNOPQRSTUVWXYZ
1234567890
Ha HaHa
MetaCharacters (Need to be escaped):
. ^ $ * + ? { } [ ] \ | ( )
coreyms.com
321-555-4321
123.555.1234
123*555*1234
800-555-1234
900-555-1234
Mr. Schafer
Mr Smith
Ms Davis
Mrs. Robinson
Mr. T
'''
```

```python
# match ----> first occurence at first index
import re

patern = re.compile(r'b')  # b is in the starting text
patern.match(text)
```

```python
# match ----> first occurence at first index
import re

patern = re.compile(r'4')  # 4 is not in the starting text
patern.match(text)
```

```python
# search ----> first occurence at any index

patern = re.compile(r'b')
patern.search(text)
```

```python
# search ----> first occurence at any index

patern = re.compile(r'4')
patern.search(text)
```

```python
# findall ----> get all ocuurance values
import re

pattern = re.compile(r'a')
pattern.findall(text) 
```

```python
## finditer  -> get all occrances and span, start,end

pattern = re.compile(r'a')
list(pattern.finditer(text))

```

```python
for i in pattern.finditer(text):
    print(i)
    print("start: ",i.start(),"Stop: ", i.end(),"span; ", i.span(),"Match: ", i.group())
```

## special characters

* .       - Any Character Except New Line
* \d      - Digit (0-9)
* \D      - Not a Digit (0-9)
* \w      - Word Character (a-z, A-Z, 0-9, _)
* \W      - Not a Word Character
* \s      - Whitespace (space, tab, newline)
* \S      - Not Whitespace (space, tab, newline)
* \b      - Word Boundary
* \B      - Not a Word Boundary
* ^       - Beginning of a String
*  dollar      - End of a String
* []      - Matches Characters in brackets
* [^ ]    - Matches Characters NOT in brackets
* |       - Either Or
* ( )     - Group

## Quantifiers:

* (*)       - 0 or More
* (+)       - 1 or More
* ?       - 0 or One
* {3}     - Exact Number
* {3,4}   - Range of Numbers (Minimum, Maximum)

```python
a = 'hello man . how are you'
pattern = re.compile(r'^h')  # ^ - starts with that index
pattern.search(a)
```

```python
a = 'This is beautiful'
pattern = re.compile(r'^b')  # ^ - starts with that index
pattern.search(a)
```

```python
a = 'hello man. how are you'

pattern = re.compile(r'u$')  # $ - ends with that index
pattern.search(a)
```
