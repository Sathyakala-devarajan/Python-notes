```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# line chart

x = np.array([0, 6])
y = np.array([0, 250])

# plt.plot(x, y)
# plt.plot(x, y, 'o')
plt.plot(y, linestyle = 'dotted', linewidth = '5', color='g') 
# linestyle = 'dashed'
# linewidth = '20.5'
# color = 'r'
plt.show()
```

```python
# subplot
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 3, 1)  # row, column, place
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 3, 3)
plt.plot(x,y)

plt.show()
```

```python
# bar chart or column chart
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

# plt.bar(x,y)
plt.bar(x, y, width = 0.1)
plt.show()
```

```python
# Horizontal bar chart
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

# plt.barh(x, y)
plt.barh(x, y, height = 0.1)
plt.show()
```

```python
# Histogram
# A histogram is a graph showing frequency distributions.

# It is a graph showing the number of observations within each given interval.
# Ex: For simplicity we use NumPy to randomly generate an array with 250 values, 
# where the values will concentrate around 170, and the standard deviation is 10.
x = np.random.normal(100, 10, 250) # looking values, std, total nos.

plt.hist(x)
plt.show() 
# print(x)
```
